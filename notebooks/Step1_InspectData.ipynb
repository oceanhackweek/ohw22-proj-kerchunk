# Seems like there isnt data for every hour each day

I wonder if there is a problem here... use dask to scan a few folders

```python
from dask import delayed, compute
from distributed import Client, LocalCluster

def check_daily_files(query_dict,url_template='s3://noaa-himawari8/{dataset}/{year}/{month}/{day}/{hour}/*{level}*.nc'):
    fs = fsspec.filesystem('s3',anon=True)
    url = url_template.format(**query_dict)
    files = fs.glob(url)
    return files

len(check_daily_files(query_dict))
```

There is going to be a fair amount of http/S3 waiting so overload a [LocalCluster](https://docs.dask.org/en/stable/deploying-python.html#reference) with a few extra threads to move things along

```python
cluster = LocalCluster(n_workers=2,threads_per_worker=4)
client = Client(cluster)
client
```

```python
daily_files={}
d_check_daily_files = delayed(check_daily_files)
for d in pd.date_range('2022-01-01','2022-05-01'):
    date = dict(year=f'{d.year}',
                month=f'{d.month:02.0f}',
                day=f'{d.day:02.0f}')
    query_dict = dict(dataset='AHI-L2-FLDK-SST',
                  hour='*',
                  level='L3C'
                 )
    query_dict.update(date)
    daily_files[d] = d_check_daily_files(query_dict)

daily_files = compute(daily_files)[0]
```

```python
# Check a couple of dates with missing files
less_than_24 = []
for d,files in daily_files.items():
    if len(files) < 24:
        less_than_24.append(d)
        display(files)
        break
```

```python
def compare_attrs(files):
    fs = s3fs.S3FileSystem(anon=True)
    ds1 = xr.open_dataset(fs.open(files[0]),engine='h5netcdf',chunks={})
    ds2 = xr.open_dataset(fs.open(files[1]),engine='h5netcdf',chunks={})
    for k in ds2.attrs.keys():
        v2 = ds2.attrs[k]
        if k not in ds1.attrs.keys():
            print(f'Missing: {k} - {v2}')
        else:
            v1 = ds1.attrs[k]
            if str(v1) != str(v2):
                print(f'Different: {k} - \n v1:{v1}\n v2:{v2}')
compare_attrs(files)
```
