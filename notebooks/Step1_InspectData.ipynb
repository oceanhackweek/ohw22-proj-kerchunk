```python
import kerchunk
import fsspec
import xarray as xr
import cf_xarray
import numpy as np
```

# Examine the source files

```python
fs = fsspec.filesystem('s3',anon=True)
```

```python
url_template = 's3://noaa-himawari8/{dataset}/{year}/{month}/{day}/{hour}/'

query_dict = dict(dataset='AHI-L2-FLDK-SST',
                  year='2022',
                  month='07',
                  day='10',
                  hour='0100')

url_base = url_template.format(**query_dict)

print(f'Files in {url_base}')
nc_files = fs.glob(f'{url_base}*.nc')
nc_files
```

```python
help(fsspec.open_files)
```

```python
for f in fsspec.open_files(nc_files,protocol='s3',anon=True):
    with f as fo:
        print(f)
        ds = xr.open_dataset(fo,engine='h5netcdf',chunks={})
        display(ds)
```

# Level 3C data looks good
